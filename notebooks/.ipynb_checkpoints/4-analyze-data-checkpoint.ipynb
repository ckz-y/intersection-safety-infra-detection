```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from tqdm import tqdm
from matplotlib.ticker import FuncFormatter
```

```python
data_path = "C:/Users/ckaz3/Desktop/intersection-safety-infra-detection/data/processed/stations_inscope.csv"
data = pd.read_csv(data_path)

station_counts = data.iloc[:, [0, 2]].groupby("Agency")["Station Name"].nunique().sort_values(ascending=False).rename("Total Stations")
mean_intersections_per_station = data.iloc[:, [0, 2, 8]].groupby(by=["Agency", "Station Name"]).count().groupby("Agency").mean().rename(columns={"Intersection_ID": "Intersections Per Station"})
result = pd.concat([station_counts, mean_intersections_per_station], axis=1, join="inner")
data["num_images"] = (data.iloc[:, 9:] == 1).sum(axis=1)
temp = data.groupby("Agency")["num_images"].mean().rename("Mean Images Per Intersection")
temp2 = data.groupby("Agency")["Intersection_ID"].nunique().rename("Total Unique Intersections")
result = pd.concat([result, temp2, temp], axis=1, join="inner")
result
```

```python
path = "C:/Users/ckaz3/Desktop/intersection-safety-infra-detection/data/interim/deployment_counts_q"
dicts=  []

for i in range(5):
    with open(path + str(i) + ".pkl", "rb") as f:
        loaded_dict = pickle.load(f)
    dicts.append(loaded_dict)
```

```python
data_path = "C:/Users/ckaz3/Desktop/intersection-safety-infra-detection/data/processed/stations_inscope.csv"
data = pd.read_csv(data_path)
data.iloc[:, 9:] = np.nan

for dict in dicts:
    for image_name in tqdm(dict):
        if "(1)" in image_name:
            continue

        island_count = dict[image_name]
        intersection_id, year = image_name.split(".")[0].split("_")
        data.loc[data["Intersection_ID"] == int(intersection_id), year] = int(island_count)
        
data
```

```python
interpolated_data = data.copy()
while interpolated_data.iloc[:, 9:].isna().sum(axis=0).sum() != 0:
    print(interpolated_data.iloc[:, 9:].isna().sum(axis=0).sum())
    interpolated_data.iloc[:, 9:] = interpolated_data.iloc[:, 9:].ffill(limit=1, axis=1).bfill(limit=1, axis=1)
```

```python
temp = (interpolated_data.iloc[:, 9:] != 0).sum(axis=0)
total_sum = len(data)

cm = 1/2.54
fig, ax1 = plt.subplots(figsize=(15*cm, 10*cm))
plt.style.use("seaborn-v0_8-paper")
ax1.grid(True)

ax1.set_xlabel("Year")
ax1.set_ylabel("Number of Intersections with a Safety Island")
ax1.plot(temp, marker='o')

ax1.set_ylim(bottom=-200, top=5200)

ax2 = ax1.twinx()

y_min, y_max = ax1.get_ylim()
ax2.set_ylim(y_min, y_max)

ax2.set_ylabel("Percentage of Intersections with a Safety Island")

def to_percent(y, position):
    return f'{100. * y / total_sum:.1f}%'

formatter = FuncFormatter(to_percent)
ax2.yaxis.set_major_formatter(formatter)

plt.tight_layout()
```

```python
temp = interpolated_data.iloc[:, 9:].sum(axis=0)

total_intersections = len(data)

cm = 1/2.54
fig, ax1 = plt.subplots(figsize=(15*cm, 10*cm))
plt.style.use("seaborn-v0_8-paper")
ax1.grid(True)

ax1.set_xlabel("Year")
ax1.set_ylabel("Number of Safety Islands")
ax1.plot(temp, marker='o')

ax1.set_ylim(bottom=-250, top=7250)

ax2 = ax1.twinx()

y_min, y_max = ax1.get_ylim()
ax2.set_ylim(y_min, y_max)

ax2.set_ylabel("Mean Safety Island per Intersection")

def to_percent(y, position):
    return f'{y / total_sum:.2f}'

formatter = FuncFormatter(to_percent)
ax2.yaxis.set_major_formatter(formatter)

plt.tight_layout()
plt.savefig("C:/Users/ckaz3/Desktop/intersection-safety-infra-detection/reports/figures/total_total.png")
```

```python
agency_abbreviations = {
    'Long Island Rail Road': 'LIRR',
    'MBTA': 'MBTA',
    'MTA': 'MTA',
    'Maryland Transit Administration': 'MDOT MTA',
    'Metro-North Railroad': 'MNR',
    'NFTA-METRO': 'NFTA-METRO',
    'NJ TRANSIT RAIL': 'NJT Rail',
    'PATCO': 'PATCO',
    "PAT ": "PRT",
    'Port Authority Trans Hudson': 'PATH',
    'SEPTA': 'SEPTA',
    'Shore Line East': 'Shore Line East',
    'VRE': 'VRE',
    'WMATA': 'WMATA'
}
```

```python
intersections_per_agency = interpolated_data.groupby("Agency").size()

agg_data = interpolated_data.groupby("Agency").sum().iloc[:, 8:]

# .sample(frac=1)

cm = 1/2.54
fig, ax = plt.subplots(figsize=(15*cm, 15*cm))

plt.style.use("seaborn-v0_8-paper")

for agency, row in agg_data.iterrows():
    # skip small agencies
    small_agencies = ["Boston Harbor Islands Ferries",
                      "Chappaquiddick Ferry,Inc.",
                      "Charm City Circulator",
                      "New England Fast Ferry",
                      "Salem Ferry"]
    second_graph = ["Maryland Transit Administration",
                    # "Metro-North Railroad",
                    "VRE",
                    "NJ TRANSIT RAIL",
                    "PAT ",
                    "PATCO",
                    "SEPTA",
                    "Shore Line East"]
    if agency in small_agencies:
        continue
    else:
        if agency in second_graph:
            ax.plot(agg_data.loc[agency] / intersections_per_agency[agency], linestyle="--", label=agency_abbreviations[agency])
        else:
            ax.plot(agg_data.loc[agency] / intersections_per_agency[agency], linestyle=None, label=agency_abbreviations[agency])
    
plt.tight_layout()
ax.grid(True)
ax.legend()
ax.set_ylabel("Mean Safety Island per Intersection")
ax.set_xlabel("Year")

plt.show()
```
