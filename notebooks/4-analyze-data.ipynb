```python
import numpy as np
import pandas as pd
```

```python
data_path = "C:/Users/ckaz3/Desktop/intersection-safety-infra-detection/data/interim/stations_inscope.csv"
data = pd.read_csv(data_path, index_col=0)
data.head()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my_app")


data[["Inter_Longitude", "Inter_Latitude"]]

def city_state_country(row):
    coord = f"{row['Inter_Latitude']}, {row['Inter_Longitude']}"
    location = geolocator.reverse(coord, exactly_one=True)
    address = location.raw['address']
    city = address.get('city', '')
    state = address.get('state', '')
    country = address.get('country', '')
    row['city'] = city
    row['state'] = state
    row['country'] = country
    return row

data.apply(city_state_country, axis=1)
```
